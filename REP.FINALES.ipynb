```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos del fichero MFINAL (separados por espacios o tabulaciones)
data = np.loadtxt('M32.txt')

# Asignar columnas a variables
x = data[:, 0]
y = data[:, 1]
yerr = data[:, 2]

# Crear la gráfica con barras de error
plt.errorbar(x, y, yerr=yerr, fmt='o', capsize=3, label='Datos', markersize=2)

plt.xlabel('T')
plt.ylabel('M')
plt.title('Representación de la magnetización en función de la temperatura')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos de los dos ficheros
# Cambia los nombres de los ficheros si es necesario
datos_positivos = np.loadtxt('densidadpositivo.txt')
datos_negativos = np.loadtxt('densidadnegativo.txt')

# Asignar columnas a variables
densidad_pos = datos_positivos[:, 0]
T_pos = datos_positivos[:, 1]

densidad_neg = datos_negativos[:, 0]
T_neg = datos_negativos[:, 1]

# Graficar ambas densidades en función de la temperatura
plt.plot(T_pos, densidad_pos, 'o-', label='Densidad positiva')
plt.plot(T_neg, densidad_neg, 's-', label='Densidad negativa')

plt.xlabel('Temperatura (T)')
plt.ylabel('Densidad')
plt.title('Densidad positiva y negativa vs Temperatura')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos del fichero (ajusta el nombre si es necesario)
datos = np.loadtxt('filecv.txt')

# Asignar columnas a variables
cv = datos[:, 0]
T = datos[:, 1]

# Graficar calor específico vs temperatura
plt.plot(T, cv, 'o', label='Calor específico')

plt.xlabel('Temperatura (T)')
plt.ylabel('Calor específico (cv)')
plt.title('Calor específico en función de la temperatura')
plt.legend()

plt.show()
```
